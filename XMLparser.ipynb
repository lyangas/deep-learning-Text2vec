```python
def log_progress(sequence, every=10):
    from ipywidgets import IntProgress
    from IPython.display import display

    progress = IntProgress(min=0, max=len(sequence), value=0)
    display(progress)
    
    for index, record in enumerate(sequence):
        if index % every == 0:
            progress.value = index
        yield record
```

```python
%%time
import xml.dom.minidom
import re
import numpy as np

#-----------------------------------------------------------
FinalRelation = np.zeros(42)
NOfTexts = 0
for i in log_progress(range (1, 4063), every=10):
    try:
        dom = xml.dom.minidom.parse('xmls/' + str(i) + '.xml');
    except Exception:
        continue
    dom.normalize()
    relation = np.zeros(42)
    for tag in dom.getElementsByTagName('tag'):
        for tagOftype in tag.childNodes[0].nodeValue.split(':'):
            try:
                for NOfTegInFinalTable in dictOfRelation[tagOftype].split(' '):
                    relation[int(NOfTegInFinalTable)] = 1
                NOfTexts = NOfTexts + 1
            except Exception:
                continue
    FinalRelation = FinalRelation + relation
    
    #-----------------------------------------------------------
    node1=dom.getElementsByTagName('source')
    string = ''
    n = 0
    while 1:
        try:
            string = string + node1[n].childNodes[0].nodeValue
            n = n+1
        except Exception:
            break

    text = string.replace('~', ' ').replace('\n', ' ')
    #-----------------------------------------------------------

    f = open ('data.txt', 'a')
    stringForFile = ''
    for k in range (42):
        stringForFile = stringForFile + str(relation[k])
        if k != 41:
            stringForFile = stringForFile + '~'
    f.write('~'+text+'~'+stringForFile+'~\n')
    f.close()
```
