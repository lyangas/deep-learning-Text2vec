```python
import torch
import torch.nn as nn
#import torchvision.datasets as dsets
#import torchvision.transforms as transforms
from torch.autograd import Variable
from IPython.display import clear_output

NOfData = 4022
input_size = 1929
num_classes = 42
```

```python
%%time
vecOfTexts = torch.zeros(input_size)#, dtype=torch.long)
vecOfResult = torch.zeros(num_classes)#, dtype=torch.long)
strings = open('data.txt' ,'r').read().split ('\n')
vec = []
for i  in range (NOfData):
    i = 0
    j = 0
    k = 0
    string = strings[i].split('|')
    
    for coordOfText in string[0].split('~'):
        vecOfTexts[i] = 10000 * float(coordOfText)
        i = i + 1
        
    for coordOfResult in string[1].split('~'):
        vecOfResult[j] = float(coordOfResult)
        j = j + 1

    vec.append([vecOfTexts, vecOfResult])
    k = k + 1

```

```python
vec_train = []
vec_test = []
for i in range (NOfData):
    if (i < 3900):
        vec_train.append(vec [i])
    else:
        vec_test.append(vec [i])
```

```python
%%time

hidden_size = 200      # Количество узлов на скрытом слое
num_epochs = 5         # Количество тренировок всего набора данных
batch_size = 1000       # Размер входных данных для одной итерации
learning_rate = 0.001  # Скорость конвергенции

class Net(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(Net, self).__init__()                    # Наследуемый родительским классом nn.Module
        self.fc1 = nn.Linear(input_size, hidden_size)  # 1й связанный слой: 784 (данные входа) -> 500 (скрытый узел)
        self.relu = nn.ReLU()                          # Нелинейный слой ReLU max(0,x)
        self.fc2 = nn.Linear(hidden_size, num_classes) # 2й связанный слой: 500 (скрытый узел) -> 10 (класс вывода)
    
    def forward(self, x):                              # Передний пропуск: складывание каждого слоя вместе
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out
    
net = Net(input_size, hidden_size, num_classes)
#net.cuda()

criterion = nn.L1Loss()#CrossEntropyLoss() LogSigmoid()
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(vec_train):   # Загрузка партии изображений с индексом, данными, классом
        images = Variable(images.view(-1, input_size))         # Конвертация тензора в переменную: изменяем изображение с вектора, размером 784 на матрицу 28 x 28
        labels = Variable(labels.view(-1, num_classes))

        optimizer.zero_grad()                             # Инициализация скрытых масс до нулей
        outputs = net(images)                             # Передний пропуск: определение выходного класса, данного изображения
        loss = criterion(outputs, labels)                 # Определение потерь: разница между выходным классом и предварительно заданной меткой
        loss.backward()                                   # Обратный проход: определение параметра weight
        optimizer.step()                                  # Оптимизатор: обновление параметров веса в скрытых узлах
        
        if (i+1) % 100 == 0:                              # Логирование
            #clear_output()
            print('Epoch [%d/%d], Step [%d/%d], Loss: %.4f'
                 %(epoch+1, num_epochs, i+1, len(vec_train)//batch_size, loss.data.item()))
            
#correct = 0
#total = 0
for images, labels in vec_test:
    images = Variable(images.view(-1, input_size))
    outputs = net(images)
    print (outputs)
    #predicted = 
    #total =                    # Увеличиваем суммарный счет
    #correct += (predicted == labels).sum()     # Увеличиваем корректный счет
 
    
#print('Accuracy of the network on the test data: %d %%' % (100 * correct / total))

#torch.save(net.state_dict(), 'fnn_model.pkl')
```
